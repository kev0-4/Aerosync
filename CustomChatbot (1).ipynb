```python
### 3. Urls and extract data from them
```

```python
URLs=[
    'https://blog.gopenai.com/paper-review-llama-2-open-foundation-and-fine-tuned-chat-models-23e539522acb',
    'https://www.mosaicml.com/blog/mpt-7b',
    'https://stability.ai/blog/stability-ai-launches-the-first-of-its-stablelm-suite-of-language-models',
    'https://lmsys.org/blog/2023-03-30-vicuna/'

]
```

```python
loaders=UnstructuredURLLoader(urls=URLs)
data=loaders.load()
```

```python
data
```

```python
len(data)
```

```python
### 4. Spliting datat into chunks
```

```python
text_splitter=CharacterTextSplitter(separator='\n',
                                    chunk_size=1000,
                                    chunk_overlap=200)
# input token limit changes with different model, change chunk size (in characters) if different mdoel
```

```python
text_chunks=text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
text_chunks[2] # Rougly 1000 characters
```

```python
### 5. Hugging Face Embeddings (Shit that would Vectorize normal words)
```

```python
#embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
embeddings=HuggingFaceEmbeddings()
```

```python
query_result = embeddings.embed_query("hello Kitty")
len(query_result)
```

```python
query_result
```

```python
### 6. Creating vector Database (Knowledgevase) from converting Chunks into Embeddings
```

```python
# FAISS is our knowledgebase (text_chunks = words , embeddings = hugging face vectorizing thing)
vectorstore=FAISS.from_documents(text_chunks, embeddings)
```

```python
### 7. Creating a LLM wrapper
```

```python
# to access model from hugging face
# hugginface.co -> settings -> access token
# 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
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2",
                                          use_auth_token=True,)

model = AutoModelForCausalLM.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
# try 8bit = False if gpu error and 4bit true, or any combination
```

```python
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
# 1 token = 4 character, changable
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
# temp changable , higher is more random
```

```python
llm.predict("give summary of the book Alchemist")
```

```python
### 8. Initialize Retrival QA with source chain data (knowledgebase(vectorstore))
```

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorstore.as_retriever())
```

```python
result=chain({"question": "How good is Vicuna?"}, return_only_outputs=True)
```

```python
result['answer']
```

```python
### 9. Contunied Conversation
```

```python
while True:
  query=input(f"Prompt: ")
  if query == 'exit':
    print('Exiting')
    sys.exit()
  if query =='':
    continue
  result=chain({'question':query})
  print(f"Answer: " +result["answer"])
```
